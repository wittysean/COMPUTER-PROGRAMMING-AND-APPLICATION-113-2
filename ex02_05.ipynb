```python
import requests
from bs4 import BeautifulSoup
import csv

def get_articles(url, cookies=None, board_name=""):
    articles = []
    response = requests.get(url, cookies=cookies)
    soup = BeautifulSoup(response.text, 'lxml')

    for atcl in soup.find_all('div', class_='r-ent'):
        if atcl.find('div', class_='title').a:
            title = atcl.find('div', class_='title').a.text.strip()
            href = 'https://www.ptt.cc' + atcl.find('div', class_='title').a.get('href')
        else:
            title = atcl.find('div', class_='title').text.strip()
            href = ''
        date = atcl.find('div', class_='date').text.strip()
        articles.append([date, href, title])
        print(f"[{board_name}] {date} | {title} | {href}")

    next_page = 'https://www.ptt.cc' + soup.find_all('a', class_='btn wide')[1].get('href')
    return next_page, articles

def scrape_board(start_url, board_name, cookies=None):
    all_articles = []
    url = start_url
    for _ in range(5):
        url, articles = get_articles(url, cookies=cookies, board_name=board_name)
        all_articles.extend(articles)
    return all_articles

def save_to_csv(filename, data):
    with open(filename, 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(['Date', 'URL', 'Title'])
        writer.writerows(data)

# 🎬 電影版
print("🎬 Scraping 電影版...")
movie_articles = scrape_board('https://www.ptt.cc/bbs/Movie/index.html', '電影')
save_to_csv('ptt_movie.csv', movie_articles)

# 🍱 美食版
print("\n🍱 Scraping 美食版...")
food_articles = scrape_board('https://www.ptt.cc/bbs/Food/index.html', '美食')
save_to_csv('ptt_food.csv', food_articles)

# 🗣️ 八卦版 (over18 cookie required)
print("\n🗣️ Scraping 八卦版...")
cookies = {'over18': '1'}
gossip_articles = scrape_board('https://www.ptt.cc/bbs/Gossiping/index.html', '八卦', cookies=cookies)
save_to_csv('ptt_gossip.csv', gossip_articles)
```
