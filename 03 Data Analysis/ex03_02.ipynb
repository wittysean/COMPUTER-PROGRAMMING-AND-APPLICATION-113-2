```python
# Step 1: Create base DataFrame
import pandas as pd
import numpy as np

data = {
    'ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack'],
    'Age': [25, 30, 35, 40, 22, 29, 38, 27, 33, 31],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia',
             'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'Salary': [50000, 60000, 75000, 80000, 45000, 62000, 72000, 58000, 69000, 71000],
    'Department': ['IT', 'HR', 'Finance', 'Marketing', 'IT', 'HR', 'Finance', 'Marketing', 'IT', 'Finance']
}
df = pd.DataFrame(data)
df
```

```python
# Step 2: Add HighSalary column (above Q3)
hs = df['Salary'].quantile(0.75)
df['HighSalary'] = df['Salary'].apply(lambda x: True if x > hs else False)
df
```

```python
# Step 3: Drop HighSalary column
df = df.drop(columns=['HighSalary'])
df
```

```python
# Step 4: Add new records using pd.concat
new_data = [
    {'ID': 111, 'Name': 'Kevin', 'Age': 28, 'City': 'Austin', 'Salary': None, 'Department': 'IT'},
    {'ID': 112, 'Name': 'Lily', 'Age': 26, 'City': 'Seattle', 'Salary': 55000, 'Department': 'Marketing'},
    {'ID': 112, 'Name': 'Lily', 'Age': 26, 'City': 'Seattle', 'Salary': 55000, 'Department': 'Marketing'}
]
new_df = pd.DataFrame(new_data)
df = pd.concat([df, new_df], ignore_index=True)
df
```

```python
# Step 5: Fill missing salaries using department averages
avg_df = df.groupby('Department')['Salary'].mean()
for i in range(len(df)):
    if pd.isnull(df.loc[i, 'Salary']):
        dept = df.loc[i, 'Department']
        df.loc[i, 'Salary'] = avg_df[dept]
df
```

```python
# Step 6: Drop duplicate rows
df = df.drop_duplicates()
df
```

```python
# Step 7: Add NewSalary (+500 if below average)
avg_salary = df['Salary'].mean()
df['NewSalary'] = df['Salary'].apply(lambda x: x + 500 if x < avg_salary else x)
df
```

```python
# Step 8: Rank salaries
df['RankSalary'] = df['Salary'].rank(ascending=False)
df
```

```python
# Step 9: Convert Salary, NewSalary, RankSalary to int
df[['Salary', 'NewSalary', 'RankSalary']] = df[['Salary', 'NewSalary', 'RankSalary']].astype(int)
df
```
