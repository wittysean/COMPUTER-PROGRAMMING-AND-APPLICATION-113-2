```python
# Step 1: Load the data from API
import pandas as pd

url = 'https://data.moenv.gov.tw/api/v2/aqx_p_432?api_key=9e565f9a-84dd-4e79-9097-d403cae1ea75&limit=1000&sort=ImportDate%20desc&format=CSV'
df = pd.read_csv(url)
df.head()
```

```python
# Step 2: Keep necessary columns
df = df[['sitename', 'county', 'pollutant', 'status', 'aqi']]
df
```

```python
# Step 3: Filter out records where AQI > 100 (poor air quality)
condition = df['aqi'] > 100
df[condition]
```

```python
# Step 4: Check all unique pollutant types
df['pollutant'].unique()
```

```python
# Step 5: Get all records for a specific city (e.g., 高雄市)
condition = df['county'] == '高雄市'
df[condition]
```

```python
# Step 6: Calculate average AQI by city (county), sorted descending
df.groupby('county')['aqi'].mean().sort_values(ascending=False)
```
